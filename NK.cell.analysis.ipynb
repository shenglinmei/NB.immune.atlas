```r
# read NK cell count matrices
raw = readRDS('NK.raw.rds')
```

```r
runConos =function (datlp2, n.cores = 8) {
    con <- Conos$new(datlp2, n.cores = n.cores)
    con$buildGraph()
    con$findCommunities()
    con$embedGraph(method = "UMAP", spread = 7)
    return(con)
    }
```

```r
p2list=lapply(raw,function(x) basicP2proc(x,n.cores = 10,min.transcripts.per.cell =0))
```

```r
# run Conos
scon = runConos(p2list)
```

```r

scon$plotGraph(size=0.2)

clu = scon$clusters$leiden$groups %>% Toch()
clu[clu=='1']='Active NK'
clu[clu=='2']='CD56dim'
clu[clu=='3']='CD56bright'
clu[clu=='4']='ILC3'

scon$plotGraph(groups = clu)
```

```r
# multilevel.community
scon$findCommunities(method = igraph::multilevel.community)
scon$plotGraph(groups = scon$clusters$`multi level`$groups,size=0.15)
```

```r
# walktrap.community

scon$findCommunities(method = igraph::walktrap.community, steps=6)
scon$plotGraph(clustering='walktrap')
```

```r
# cluster stability
scon$findCommunities(method=leiden.community, resolution=1,test.stability = TRUE)
boxplot(scon$clusters$leiden$stability$flat$jc[,1:4],ylab='Jaccard index')
```

```r
fraction.palette1 = c('grey','yellow','purple','red')
names(fraction.palette1) = c('Adrenal','Low risk','Intermediate risk','High risk')
```

```r
ssamp = scon$getDatasetPerCell()
cname=names(anoM)
ano2=data.frame('Cell'=anoM[cname],'SampleType'=ssamp[cname])

# Annotation vs sample
tmp2 <- acast(ano2, Cell ~ SampleType, fun.aggregate=length)

tmp3 <- (sweep(tmp2, 2, colSums(tmp2), FUN='/'))
tmp4 <- melt(tmp3)
head(tmp4)

names(tmp4) <- c('cell', 'sample','pc.of.sample')

tmp4$Group=NULL
tmp4$Group=sample.groups2[as.character(tmp4$sample)]


p <- ggplot(na.omit(tmp4),aes(x=cell,y=pc.of.sample,dodge=Group,fill=Group))+geom_boxplot(notch=FALSE,outlier.shape=NA)  +  geom_point(position = position_jitterdodge(jitter.width=0.1),color=adjustcolor(1,alpha=0.3),pch=19,size=0.5)+theme_classic()+
  theme(axis.text.x = element_text(angle = 90, hjust = 1), axis.text.y = element_text(angle = 90, hjust = 0.5))  +xlab("") +ylab("fraction of total NK cells")+theme(legend.position="right")+
  scale_fill_manual(values=fraction.palette1[2:4])
p

#ggsave('NK.fraction.pdf',p,height=2.8,width=4.1)
```

```r

library(ggpubr)

df=tmp4
rsig=NULL
for (i in unique(df[,1])){
  tmp=df[df[,1]==i,]

  sig=compare_means(pc.of.sample ~ Group,  data = tmp) # 
  sig$cell=i
  rsig=rbind(rsig,sig)
  #sig[sig$p.signif!='ns',]
}

write.table(rsig,'NK.fraction.1.pvalue.xls',sep='\t',col.names=T,row.names=F,quote=F)
```
